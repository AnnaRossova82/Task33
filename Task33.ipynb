```python
df=pd.read_excel('E:\Cars.xlsx')
arr=df.to_numpy()
print(arr)
              
```

```python
car_brand=df.loc[:,'car_brand'].values
cost=df.loc[:,'cost'].values
fuel_use=df.loc[:,'fuel_use'].values
max_speed=df.loc[:,'max_speed'].values
load_capacity=df.loc[:,'load_capacity'].values
```

```python
#Залежність споживання палива від вартості авто
fig, ax=plt.subplots()
plt.style.use('classic')
plt.plot(fuel_use,cost, c='r')
plt.xlabel('USD')
plt.ylabel('Volume of use')
plt.title('How the fuel use depends on car cost')
plt(show)
```

```python
#Залежність споживання палива від вартості авто, друга спроба просто для цікавості
plt.style.use('seaborn')
fig, axes=plt.subplots(nrows=1,ncols=1)
axes.plot(cost, fuel_use, c='b')
```
